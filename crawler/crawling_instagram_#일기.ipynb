```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
import csv 
```

```python
key = ['일기']

# 나중에 결과 저장할 딕셔너리 생성
insta_dict = {}

# selenium 이용해서 드라이버 지정하고 url 불러오기
# keyword 를 포함한 url로 쓰기
driver = webdriver.Chrome('./chromedriver.exe') 

#keyword를 for문으로 돌릴 수 있으나 나중에 driver에서 오류가 날 가능성이 높음.
keyword = key[0]
insta_url = "https://www.instagram.com"
search_url = insta_url + "/explore/tags/" + keyword
driver.get(search_url)
driver.maximize_window() #창 크게
time.sleep(3)


## 로그인
# 아이디, 비번 변수에 저장
insta_id = "아이디"
insta_pw = "비밀번호"


driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/span/a[1]/button').click()
#print(find_login) //*[@id="loginForm"]/div/div[1]/div/label/input //*[@id="loginForm"]/div/div[2]/div/label/input

time.sleep(5)

# id 입력
input_id = driver.find_element_by_xpath('//*[@id="loginForm"]/div/div[1]/div/label/input')
#input_id.click()
input_id.send_keys(insta_id)

# pw 입력
input_password = driver.find_element_by_xpath('//*[@id="loginForm"]/div/div[2]/div/label/input')
#input_password.click()
input_password.send_keys(insta_pw)

# 로그인 버튼 클릭 click
login = driver.find_element_by_css_selector("button.sqdOP.L3NKy.y3zKF")
login.click()

## 로그인 시 팝업 창 제거
# 로그인 저장 팝업 제거
time.sleep(5)
driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/div/div/div/button').click()

# 알림창 제거
'''time.sleep(10)
alert = driver.find_elements_by_css_selector("button.aOOlW")
alertLater = alert
alertLater.click()'''
```

```python
## 첫번째 사진 클릭 후 텍스트 가져오기~다음버튼 클릭하기 반복하기
# 첫번째 사진 클릭하기
driver.find_element_by_xpath('//*[@id="react-root"]/section/main/article/div[1]/div/div/div[1]/div[1]/a/div/div[2]').click()

#크롤링한 결과 넣을 리스트, 그거 모을 딕트 생성
texts = []


#반복문. 10단위로 끝날때마다 알림프린트
i = 0
text = driver.find_element_by_xpath('/html/body/div[4]/div[2]/div/article/div/div[3]/div[1]/ul/div/li/div/div/div[2]/span').text
texts.append(text)
time.sleep(1)
driver.find_element_by_xpath('/html/body/div[4]/div[1]/div/div/a').click()
time.sleep(2)

while i < 100000:
    i += 1
    # # 게시물 텍스트 가져오기
    try:
        text = driver.find_element_by_xpath('/html/body/div[4]/div[2]/div/article/div/div[3]/div[1]/ul/div/li/div/div/div[2]/span').text
        texts.append(text)
    except:
        pass
    time.sleep(1)
    # 다음 버튼 클릭하기
    driver.find_element_by_xpath('/html/body/div[4]/div[1]/div/div/a[2]').click()
    time.sleep(2)
    if i % 10 == 0:
        print("{0}번 끝".format(i))
```

```python
# 결과 dict에 저장하기
insta_dict[ keyword ] = texts

# 창 나가기
actions = webdriver.ActionChains(driver).send_keys(Keys.ESCAPE)
actions.perform()


# 완성 딕셔너리 확인
insta_dict

# 딕셔너리를 csv 파일에 저장하기
with open('insta_texts.csv', 'w', encoding='utf-8', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = ['index','city','insta_text'])
    writer.writeheader()
    for key, value in insta_dict.items():
        i = 0
        writer.writerow({'index': i , 'city': key, 'insta_text': value})
        i += 1
```

```python
from unicodedata import normalize
```

```python
with open('instagram_#일기_7519.xlsx','w',encoding='utf-8') as f:
    for i in range(len(texts)):
        t = texts[i].replace('\n',' ')#.replace('#',' ')
        t = normalize('NFC', t)
        f.write(t+'\n')
```
