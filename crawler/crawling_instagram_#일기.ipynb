```python
## ì²«ë²ˆì§¸ ì‚¬ì§„ í´ë¦­ í›„ í…ìŠ¤íŠ¸ ê°€ì ¸ì˜¤ê¸°~ë‹¤ìŒë²„íŠ¼ í´ë¦­í•˜ê¸° ë°˜ë³µí•˜ê¸°
# ì²«ë²ˆì§¸ ì‚¬ì§„ í´ë¦­í•˜ê¸°
driver.find_element_by_xpath('//*[@id="react-root"]/section/main/article/div[1]/div/div/div[1]/div[1]/a/div/div[2]').click()

#í¬ë¡¤ë§í•œ ê²°ê³¼ ë„£ì„ ë¦¬ìŠ¤íŠ¸, ê·¸ê±° ëª¨ì„ ë”•íŠ¸ ìƒì„±
texts = []


#ë°˜ë³µë¬¸. 10ë‹¨ìœ„ë¡œ ëë‚ ë•Œë§ˆë‹¤ ì•Œë¦¼í”„ë¦°íŠ¸
i = 0
text = driver.find_element_by_xpath('/html/body/div[4]/div[2]/div/article/div/div[3]/div[1]/ul/div/li/div/div/div[2]/span').text
texts.append(text)
time.sleep(1)
driver.find_element_by_xpath('/html/body/div[4]/div[1]/div/div/a').click()
time.sleep(2)

while i < 100000:
    i += 1
    # # ê²Œì‹œë¬¼ í…ìŠ¤íŠ¸ ê°€ì ¸ì˜¤ê¸°
    try:
        text = driver.find_element_by_xpath('/html/body/div[4]/div[2]/div/article/div/div[3]/div[1]/ul/div/li/div/div/div[2]/span').text
        texts.append(text)
    except:
        pass
    time.sleep(1)
    # ë‹¤ìŒ ë²„íŠ¼ í´ë¦­í•˜ê¸°
    driver.find_element_by_xpath('/html/body/div[4]/div[1]/div/div/a[2]').click()
    time.sleep(2)
    if i % 10 == 0:
        print("{0}ë²ˆ ë".format(i))
```

```python
# ê²°ê³¼ dictì— ì €ì¥í•˜ê¸°
insta_dict[ keyword ] = texts

# ì°½ ë‚˜ê°€ê¸°
actions = webdriver.ActionChains(driver).send_keys(Keys.ESCAPE)
actions.perform()


# ì™„ì„± ë”•ì…”ë„ˆë¦¬ í™•ì¸
insta_dict

# ë”•ì…”ë„ˆë¦¬ë¥¼ csv íŒŒì¼ì— ì €ì¥í•˜ê¸°
with open('insta_texts.csv', 'w', encoding='utf-8', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = ['index','city','insta_text'])
    writer.writeheader()
    for key, value in insta_dict.items():
        i = 0
        writer.writerow({'index': i , 'city': key, 'insta_text': value})
        i += 1
```
