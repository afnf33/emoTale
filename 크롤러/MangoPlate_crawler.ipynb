```python
import requests
import json
import time
from bs4 import BeautifulSoup
from urllib import parse
```

```python
def get_restaurants_url_list(soup):
    back_url_list = []
    list_restaurants = soup.find_all('div', class_='list-restaurant-item')
    for restaurant in list_restaurants:
        restaurant_href = restaurant.find('div', class_='info').find('a')['href']
        back_url_list.append(restaurant_href)
    return back_url_list
```

```python
def get_review(back_url_list):
    comments = []
    custom_header = {'referer': "https://www.mangoplate.com/",
                    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36'}
    front_url = "https://www.mangoplate.com"
    for back_url in back_url_list:
        #api url과 기본 url이 다름에 주의
        url = f"https://stage.mangoplate.com/api/v5{back_url}/reviews.json?language=kor&device_uuid=XtKwD15955102691932774pg7tV&device_type=web&start_index=0&request_count=5&sort_by=2"
        req = requests.get(url, headers = custom_header)
        if req.status_code == requests.codes.ok:
            print('접속 성공')
            list_comments = json.loads(req.text)
            for comment in list_comments:
                review = comment['comment']
                comments.append(review['comment'].replace('\n', ''))
        else:
            print("접속 실패")
    return comments
```

```python
def get_search_soup(search, i):
    restaurant_list=[]
    #custom_header가 없을 경우 컴퓨터로 인식하고 거부할 수도....
    custom_header = {'referer': "https://www.mangoplate.com/",
                    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36'}
    url = f"https://www.mangoplate.com/search/{search}?keyword={search}&page={i}"
    req = requests.get(url, headers = custom_header)
    soup = BeautifulSoup(req.text, "html.parser")
    return soup
```

```python
def main():
    f = open('test.txt', 'w', encoding='utf8')
    search_list = []
    for i in range(10):
        search = input("검색어를 입력하세요: ")
        if search in search_list:
            continue
        else:
            review_list = []
            for i in range(1, 11):
                soup = get_search_soup(search, i)
                back_url_list = get_restaurants_url_list(soup)
                current_review_list = get_review(back_url_list)
                review_list.append(current_review_list)
            result = []
            for i in range(len(review_list)):
                for review in review_list[i]:
                    f.write(review+'\n\n') 
    f.close()
```

```python
if __name__ == "__main__":
    main()
```
